```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    lines = [line.split(" ") for line in lines]
    return lines
```

```python
machines = get_data("input_test.txt")
lights = [list(machine[0][1:-1]) for machine in machines]
buttons = [[list(map(int, item[1:-1].split(","))) for item in machine[1:-1]] for machine in machines]
jolt_reqs = [eval(machine[-1]) for machine in machines]
```

```python
for m, machine in enumerate(machines):
    m_lights = lights[m]
    m_buttons = buttons[m]
    m_jolt_reqs = jolt_reqs[m]

    start_lights = len(m_lights)*["."]
    button_combs = m_buttons

    print(start_lights, m_lights, m_buttons, m_jolt_reqs)
```
