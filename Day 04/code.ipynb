```python
from itertools import *
```

```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    grid = [list(line) for line in lines]
    return grid
```

```python
def get_roll_counts(grid):
    roll_counts = []
    for r, row in enumerate(grid):
        for c, col in enumerate(row):
            val = grid[r][c]
            if val == "@":
                adj_val_counts = {"@":0, ".":0, "x":0}
                for r_, c_ in list(product([-1, 0, 1], repeat=2)):
                    r_offset = r + r_
                    c_offset = c + c_
                    if r_offset >= 0 and c_offset >= 0 and r_offset < len(grid) and c_offset < len(grid[0]):
                        val_offset = grid[r_offset][c_offset]
                        adj_val_counts[val_offset] += 1
                roll_counts.append([(r, c), adj_val_counts])
    return roll_counts

def get_accessible_positions(roll_counts):
    accessible_positions = []
    for pos, counts in roll_counts:
        adj_roll_count = counts["@"]
        if adj_roll_count <= 4:
            accessible_positions.append(pos)
    return accessible_positions

def get_result(grid):
    roll_counts = get_roll_counts(grid)
    accessible_positions = get_accessible_positions(roll_counts)
    return accessible_positions

grid = get_data("input.txt")
accessible_positions = get_result(grid)
print(len(accessible_positions))
```

Part 2

```python
grid = get_data("input.txt")
total_rolls = 0
while True:
    accessible_positions = get_result(grid)
    for r, c in accessible_positions:
        grid[r][c] = "x"
    total_rolls += len(accessible_positions)
    if len(accessible_positions) == 0:
        break
print(total_rolls)
```
