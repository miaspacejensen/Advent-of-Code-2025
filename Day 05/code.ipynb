```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    sections = text.split("\n\n")
    lines = [lines.split("\n") for lines in sections]
    ranges = [list(map(int, item.split("-"))) for item in lines[0]]
    ids = [int(item) for item in lines[1]]
    return ranges, ids
```

```python
def check_in_range(id, id_ranges):
    in_ranges = []
    for id_range in id_ranges:
        if id >= id_range[0] and id <= id_range[1]:
            in_ranges.append(id_range)
    return in_ranges
```

```python
def get_fresh_count(id_ranges, ids):
    fresh_count = 0
    for id in ids:
        in_ranges = check_in_range(id, id_ranges)
        if in_ranges != []:
            fresh_count += 1
    return fresh_count
```

```python
id_ranges, ids = get_data("input.txt")
get_fresh_count(id_ranges, ids)
```

```python
def count_items_in_ranges(id_ranges):
    sum_diff = 0
    for lower, upper in id_ranges:
        diff = (upper + 1) - lower
        sum_diff += diff 
    return sum_diff
```

```python
id_ranges, ids = get_data("input.txt")
id_ranges.sort(key=lambda x: (x[0], x[1]))

id_ranges_new = id_ranges

j = 1
while True:
    if j == len(id_ranges_new):
        break

    i = j-1

    id_range_j = lower_j, upper_j = id_ranges_new[j]
    id_range_i = lower_i, upper_i = id_ranges_new[i]

    if upper_i >= lower_j and upper_i <= upper_j: # Overlapping
        id_range_new = [id_range_i[0], id_range_j[1]]
        id_ranges_new = id_ranges_new[:i] + [id_range_new] + id_ranges_new[i+2:]    
    elif upper_i >= lower_j and upper_i > upper_j: # Encompassing
        id_ranges_new = [element for i, element in enumerate(id_ranges_new) if i != j]
    else:
        j += 1

count_items_in_ranges(id_ranges_new)
```
