```python
import itertools
```

```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    banks = text.split("\n")
    banks = [list(map(int, list(bank))) for bank in banks]
    return banks
```

```python
def get_result(banks):
    total_joltage = 0
    for bank in banks:
        largest_battery = 0
        for i, battery_start in enumerate(bank):
            options = bank[i+1:]
            if options != []:
                battery_end = max(options)
                battery = int(str(battery_start) + str(battery_end))
            if battery > largest_battery:
                largest_battery = battery
        total_joltage += largest_battery
    return total_joltage
```

```python
banks = get_data("input.txt")
get_result(banks)
```

```python
def get_result(banks):
    total_joltage = 0
    for bank in banks:
        largest_battery = 0
        for i in list(range(len(bank)))[:len(bank)-11]:
            battery_start = bank[i]
            remaining = bank[i+1:]
            build_battery = [battery_start]
            j = i
            k = 0
            while True:
                options = bank[j+1:len(bank)-10+k]
                if len(build_battery) + len(remaining) == 12 or options == []:
                    build_battery += remaining
                    break
                elif len(build_battery) == 12:  
                    break
                selected_option = max(options)
                j += options.index(selected_option) + 1
                build_battery.append(selected_option)
                k+=1
                remaining = bank[j+1:]
            battery = int(''.join(str(num) for num in build_battery))
            if battery > largest_battery:
                largest_battery = battery
        total_joltage += largest_battery
    return total_joltage
```

```python
banks = get_data("input.txt") 
get_result(banks) 
```
