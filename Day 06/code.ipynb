```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    worksheet = [line.split() for line in lines]
    return worksheet 
```

```python
def get_total(worksheet, show_problems=False):
    total = 0
    ind = 0
    for _ in range(len(worksheet[0])):
        items = [problem[ind] for problem in worksheet]
        vals = items[:len(worksheet)-1]
        opp = " "+items[-1]+" "
        problem = opp.join(vals)
        solution = eval(problem)
        if show_problems: print(problem, "=", solution)
        total += solution
        ind += 1
    return total
```

```python
worksheet1 = get_data("input.txt")
get_total(worksheet1)
```

```python
def get_data2(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    worksheet = [list(line) for line in lines]
    return worksheet
```

```python
worksheet2 = get_data2("input.txt")
```

```python
current_worksheet = worksheet2
grand_total = 0
while True:
    i = 0
    problem = []
    while True:
        items = [problem[i] for problem in current_worksheet]
        if i == 0:
            opp = " "+items[-1]+" "
        vals = items[:len(worksheet2)-1]
        vals = "".join([item for item in vals if item != " "])
        if (i == len(current_worksheet[0])-1):
            problem += [vals]
        if (vals == "") or (i == len(current_worksheet[0])-1):
            full_problem = opp.join(problem)
            solution = eval(full_problem)
            grand_total += solution
            current_worksheet = [problem[i+1:] for problem in current_worksheet]
            break
        problem += [vals]
        i += 1
    if all(item == [] for item in current_worksheet):
        break
grand_total
```
