```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    grid = [list(line) for line in lines]
    return grid 
```

```python
def display_grid(grid):
    for row in grid:
        print("".join(row)) 
```

```python
def get_start_pos(grid, star_val="S"):
    for r, row in enumerate(grid):
        for c, col in enumerate(row):
            val = grid[r][c]
            if val == star_val:
                return (r, c)
```

```python
def get_next_pos(grid, curr_pos):
    curr_r, curr_c = curr_pos
    next_pos = next_r, next_c = curr_r + 1, curr_c
    next_val = grid[next_r][next_c]
    if next_val == ".":
        next_pos = [(next_r, next_c)] = [(curr_r + 1, curr_c)]
    elif next_val == "^":
        next_pos = [(next_r, next_c - 1), (next_r, next_c + 1)]
    return next_pos
```

```python
def get_beam_path(grid, show_grid=True):
    curr_coords = [get_start_pos(grid, star_val="S")]
    beam_coords = []
    num_splits = 0
    i = 0
    while True:
        if i == len(grid)-1:
            break
        curr_coords = [pos for pos in curr_coords if (pos[1] < len(grid[0])) and (pos[1] >= 0)]
        next_coords = []
        for pos in curr_coords:
            next_coord = get_next_pos(grid, pos)
            if len(next_coord) > 1:
                num_splits += 1
            next_coords += next_coord
        curr_coords = list(set(next_coords))
        beam_coords += curr_coords
        i += 1
    grid_updated = grid.copy()
    for r, c in beam_coords:
        grid_updated[r][c] = "|"
    if show_grid: display_grid(grid_updated)
    return beam_coords, num_splits

grid = get_data("input.txt")
beam_coords, num_splits = get_beam_path(grid, show_grid=True)
num_splits
```
