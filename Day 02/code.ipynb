```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    ranges = text.split(",")
    ranges = [list(map(int, range.split("-"))) for range in ranges]
    return ranges
```

```python
def get_result(ranges):
    invalid_ids = {}
    sum_invalid_ids = 0
    for i, (first_id, last_id) in enumerate(ranges):
        id_range = list(range(first_id, last_id+1))
        for id in id_range:
            id_str = str(id)
            len_id = len(id_str)
            if len_id % 2 == 0:
                halfway = int(len_id/2)
                first_half = id_str[:halfway]
                second_half = id_str[halfway:]
                if first_half == second_half:
                    sum_invalid_ids += id
                    if i not in invalid_ids:
                        invalid_ids[i] = [id]
                    else:
                        invalid_ids[i].append(id)
    return invalid_ids, sum_invalid_ids
```

```python
ranges = get_data("input_test.txt")
invalid_ids, sum_invalid_ids = get_result(ranges)
sum_invalid_ids
```

```python
def check_all_equal(num_list):
    return all(num == num_list[0] for num in num_list)
```

```python
def sum_lists_of_invalid_ids(dic):
    vallist = []
    for _, val in dic.items():
        vallist += val
    return sum(vallist)
```

```python
def get_result(ranges):

    invalid_ids = {}

    for i, (first_id, last_id) in enumerate(ranges):

        id_range = list(range(first_id, last_id+1))

        for id in id_range:

            id_str = str(id)
            len_id = len(id_str)
            len_range = list(range(1, len_id))
            
            valid_chunk_sizes = [div for div in len_range if len_id % div == 0]

            for chunk_size in valid_chunk_sizes:

                chunks = [id_str[j:j+chunk_size] for j in range(0, len_id, chunk_size)]

                if check_all_equal(chunks):

                    if i not in invalid_ids:
                        invalid_ids[i] = [id]

                    elif id not in invalid_ids[i]:
                        invalid_ids[i].append(id)

    sum_invalid_ids = sum_lists_of_invalid_ids(invalid_ids)

    return invalid_ids, sum_invalid_ids
```

```python
ranges = get_data("input.txt")
invalid_ids, sum_invalid_ids = get_result(ranges)
sum_invalid_ids 
```
