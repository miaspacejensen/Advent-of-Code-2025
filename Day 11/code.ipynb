```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    lines = [line.split(": ") for line in lines]
    lines = [[line[0], line[1].split(" ")] for line in lines]
    graph = dict(lines)
    return graph
```

```python
def get_graph_paths(graph, node_in, node_out):
    visited_nodes_list = []
    def dfs(graph, curr_node, visited_nodes):
        visited_nodes.append(curr_node)
        for node in graph[curr_node]:
            if graph[curr_node][0] == node_out:
                visited_nodes_list.append(visited_nodes + graph[curr_node])
                break
            if node not in visited_nodes:
                dfs(graph, node, visited_nodes.copy())
    dfs(graph, node_in, [])
    return visited_nodes_list
```

```python
graph = get_data("input.txt")
all_paths = get_graph_paths(graph, node_in="you", node_out="out")
len(all_paths)
```

```python
def get_graph_paths2(graph, node_in, node_out, must_visit_nodes=["dac", "fft"]):
    visited_nodes_list = []
    def dfs(graph, curr_node, visited_nodes):
        visited_nodes.append(curr_node)
        for node in graph[curr_node]:
            if graph[curr_node][0] == node_out:
                if all(node in visited_nodes for node in must_visit_nodes):
                    visited_nodes_list.append(visited_nodes + graph[curr_node])
                break
            if node not in visited_nodes:
                dfs(graph, node, visited_nodes.copy())
    dfs(graph, node_in, [])
    return visited_nodes_list
```

```python
graph = get_data("input.txt")
all_paths = get_graph_paths2(graph, node_in="svr", node_out="out")
len(all_paths)
```
