```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    pairs = [line.split(",") for line in lines]
    pairs = [[int(num) for num in pair] for pair in pairs]
    return pairs 
```

```python
def get_areas(red_tiles):
    largest_area = [[], [], 0]
    areas = []
    i = 0
    while True:
        red_tile1 = red_tiles[i]
        for red_tile2 in red_tiles[i+1:]:
            height = abs((red_tile2[1] - red_tile1[1])) + 1
            width = abs((red_tile2[0] - red_tile1[0])) + 1
            area = height * width
            area_set = [red_tile1, red_tile2, area]
            areas += [area_set]
            if area > largest_area[2]:
                largest_area = area_set
        i += 1
        if i == len(red_tiles):
            break
    return areas, largest_area
```

```python
red_tiles = get_data("input.txt")
areas, largest_area = get_areas(red_tiles)
largest_area[2]
```

```python
def create_grid(red_tiles):
    max_row = max(row for col, row in red_tiles)
    max_col = max(col for col, row in red_tiles)
    grid = [["." for _ in range(max_col+3)] for _ in range(max_row+2)]
    return grid
```

```python
def display_grid(grid):
    for row in grid:
        print("".join(row))
```

```python
def get_loop(red_tiles):
    i = 0
    green_edge_tiles = []
    edges = []
    while True:

        red_tile1 = red_tiles[i]

        for red_tile2 in red_tiles[i+1:]:

            green_tiles = []

            if red_tile1[0] == red_tile2[0]:
                edge = [red_tile1, red_tile2]
                edges += [edge]
                if red_tile1[1] > red_tile2[1]:
                    step = -1
                else:
                    step = 1
                col_positions = list(range(red_tile1[1]+step, red_tile2[1], step))
                for col in col_positions:
                    green_tiles += [[red_tile1[0], col]]
                # print("H", [red_tile1] + green_tiles + [red_tile2], col_positions)

            elif red_tile1[1] == red_tile2[1]:
                edge = [red_tile1, red_tile2]
                edges += [edge]
                if red_tile1[0] > red_tile2[0]:
                    step = -1
                else:
                    step = 1
                row_positions = list(range(red_tile1[0]+step, red_tile2[0], step))
                for row in row_positions:
                    green_tiles += [[row, red_tile1[1]]]
                # print("V", [red_tile1] + green_tiles + [red_tile2], row_positions)
        
            green_edge_tiles += green_tiles
        
        i += 1
        if i == len(red_tiles):
            break

    return green_edge_tiles, edges
```

```python
def get_tiles_inside_loop(edges, loop_tiles):
    
    row_min = min(row for col, row in loop_tiles)
    row_max = max(row for col, row in loop_tiles)
    
    all_tiles = []
    
    for row in range(row_min, row_max + 1):

        intersections = []

        for edge in edges:
            (c1, r1), (c2, r2) = edge
            if (r1 != r2) and (row >= min(r1, r2)) and (row <= max(r1, r2)):
                col_intersect = c1 + (row - r1) * (c2 - c1) / (r2 - r1)
                intersections.append(col_intersect)

        intersections.sort()

        for i in range(0, len(intersections), 2):
            if i + 1 < len(intersections):
                col_start, col_end = int(intersections[i]), int(intersections[i+1])
                all_tiles.extend([[col, row] for col in range(col_start, col_end + 1)])

    return all_tiles
```

```python
red_tiles = get_data("input_test.txt")
green_edge_tiles, edges = get_loop(red_tiles)
loop_tiles = red_tiles + green_edge_tiles
all_tiles = get_tiles_inside_loop(edges, loop_tiles)
```

```python
grid = create_grid(all_tiles)
for c, r in all_tiles:
    if [c, r] in red_tiles:
        grid[r][c] = "#"
    else:
        grid[r][c] = "X"
display_grid(grid)
```
