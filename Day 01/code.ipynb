Part 1

```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    rotations = text.split("\n")
    pairs = [[r[0], int(r[1:])] for r in rotations] 
    return pairs
```

```python
def get_zero_count(rotations, position=50, show_steps=True):

    position += 1

    if show_steps==True:
        print(f"The dial starts by pointing at {position-1}.")

    count = 0

    for dir, steps in rotations:

        if dir == "L":
            pos = position - steps
        elif dir == "R":
            pos = position + steps

        if pos < 1: 
            position = 100 - (-pos % 100)
        elif pos > 100 :
            position = (pos % 100)
        else:
            position = pos

        if position == 1:
            count += 1

        if show_steps==True:
            print(f"The dial is rotated {dir}{steps} to point at {position-1}.")

    return count
```

```python
rotations = get_data("input.txt")
get_zero_count(rotations, position=50, show_steps=False) 
```
