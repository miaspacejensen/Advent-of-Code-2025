```python
def get_data(file_name):
    with open(file_name, "r") as file:
        text = file.read()
    lines = text.split("\n")
    boxes = [line.split(",") for line in lines]
    boxes = [[int(num) for num in box] for box in boxes]
    return boxes 
```

```python
box_positions = get_data("input_test.txt")
box_positions
```

```python
def get_distance(coord1, coord2):
    x1, y1, z1 = coord1
    x2, y2, z2 = coord2
    d = sqrt((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)
    return d
```

```python
def get_closest_coords(box_positions):
    i = 0
    all_distances = []
    closest_distance = 0
    closest_coords = []
    while True:
        if i == len(box_positions):
            break
        coord1 = box_positions[i]
        for coord2 in box_positions[i+1:]:
            d = get_distance(coord1, coord2)
            all_distances += [[[coord1, coord2], d]]
            if (closest_distance == 0) or (d<closest_distance):
                closest_distance = d
                closest_coords = [coord1, coord2]
        i += 1
    return all_distances, closest_coords, closest_distance
```
